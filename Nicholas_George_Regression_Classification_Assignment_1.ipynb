```python
# Read New York City apartment rental listing data
import pandas as pd
df = pd.read_csv(DATA_PATH+'apartments/renthop-nyc.csv')
assert df.shape == (49352, 34)

df.head()
```

```python
# Remove outliers: 
# the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
df = df[(df['price'] >= 1375) & (df['price'] <= 15500) & 
        (df['latitude'] >=40.57) & (df['latitude'] < 40.99) &
        (df['longitude'] >= -74.1) & (df['longitude'] <= -73.38)]
```

```python
# Step 1
from sklearn.linear_model import LinearRegression
```

```python
#initiate the class
model = LinearRegression()
```

```python
#arranging x features matrix
features = ['bedrooms']

#y target vector
target = 'price'
x = df[features]
y = df[target]

print(x.shape, y.shape)
```

```python
#fitting the model
model.fit(x,y)
```

```python
#apply the model to new data
bedrooms = 4
x_test = [[bedrooms]]
y_pred = model.predict(x_test)

print(f'Predicted monthly price for {bedrooms} bedroom apartment New York City condo: ${round(y_pred[0],2)}.')
```

```python
#finding the error
y_test = [5120]

from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print(f'Our models error from a random guess is: ${round(mae,2)}.')
```

```python
preds = model.predict(x)
mae = mean_absolute_error(y, preds)
print(f'Our models MAE from all the data is: ${round(mae,2)}')
```

```python
#comparing it to the mean guess, our model is better
guess = df['price'].mean()
errors = guess - df['price']
mean_absolute_error = errors.abs().mean()
print(f'If we just guessed every apartment rented for ${guess:,.0f},')
print(f'we would be off by ${mean_absolute_error:,.0f} on average.')
```

```python
#comparing it to the median
guess = df['price'].median()
errors = guess - df['price']
mean_absolute_error = errors.abs().mean()
print(f'If we just guessed every apartment rented for ${guess:,.0f},')
print(f'we would be off by ${mean_absolute_error:,.0f} on average.')
```
